# RGB

```python
(b,g,r) = cv2.split(img)
```

Se muestran los canales con MatPlotLib

```python
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(r, cmap='gray')
plt.title('R CHANNEL')
plt.subplot(132)
plt.imshow(g, cmap='gray')
plt.title('G CHANNEL')
plt.subplot(133)
plt.imshow(b, cmap='gray')
plt.title('B CHANNEL')
plt.show()
```

# CLASE: TRANSFORMACION A ESCALA DE GRISES

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
img_gray.shape
```

RGB ---> a una escala de grises ponderada

```python
# LUMINICENCIA
w_r = 0.299
w_g = 0.587
w_b = 0.114
```

```python
w_r = 0.5
w_g = 0.5
w_b = 0
```

```python
gray_lum = w_r * r + w_g * g + w_b * b
```

```python
#IMG_1
plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(img_gray, cmap='gray')
plt.title('Img escala de grises promedio')
#IMG_2
plt.subplot(122)
plt.imshow(gray_lum, cmap='gray')
plt.title('Img escala de grises ponderado')


plt.show()
```

# HSV: Transformaciones sobre color, sin alterar brillo

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
(h, s, v) = cv2.split(img_hsv)
```

```python
x, y = 0, 0
print(f"H: p[{y},{x}] = {h[y,x]}")
print(f"S: p[{y},{x}] = {s[y,x]}")
print(f"V: p[{y},{x}] = {v[y,x]}")
```

```python
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(h, cmap='gray')
plt.title('H CHANNEL - TONALIDAD')
plt.subplot(132)
plt.imshow(s, cmap='gray')
plt.title('S CHANNEL - SATURACION(INTENSIDAD DEL COLOR) ')
plt.subplot(133)
plt.imshow(v, cmap='gray')
plt.title('V CHANNEL - BRILLO')
plt.show()
```

```python
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
(h_2, l , s_2) = cv2.split(img_hls)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(h_2, cmap='gray')
plt.title('H CHANNEL - TONALIDAD')
plt.subplot(132)
plt.imshow(l, cmap='gray')
plt.title('L CHANNEL - LUMINOSIDAD')
plt.subplot(133)
plt.imshow(s_2, cmap='gray')
plt.title('S CHANNEL - SATURARION')
plt.show()
```
